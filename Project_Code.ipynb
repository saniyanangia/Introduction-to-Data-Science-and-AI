# **Northeast region analysis**

```python
covidData_Northeast = covidData.loc[covidData.state == 'Northeast', :]
covidData_Northeast.head()
```

```python
date = pd.DataFrame(covidData_Northeast['date'])  
pos_per_capita_Northeast = pd.DataFrame(covidData_Northeast['pos_per_capita']) 
```

```python
covidData_Northeast = pd.concat([date, pos_per_capita_Northeast], axis = 1).reindex(date.index)
covidData_Northeast.head()
```

```python
covidData_Northeast.isnull().sum()  # checking for null values
```

```python
def plot_df(covidData_Northeast, x, y, title = "", xlabel = 'Date', ylabel = 'Case Count Per Capita', dpi = 100):
    plt.figure(figsize = (20,5), dpi = dpi)
    plt.plot(x, y, color = 'tab:green')
    plt.gca().set(title = title, xlabel = xlabel, ylabel = ylabel)
    plt.show()

plot_df(covidData_Northeast, x = covidData_Northeast.date, y = covidData_Northeast.pos_per_capita, title = 'Northeast Cases Per Capita')
```

#### This time series has an increasing trend, but not seasonality

```python
#ADF Test
from statsmodels.tsa.stattools import adfuller, kpss #testing for stationary 
result = adfuller(pos_per_capita_Northeast.values, autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'{key}, {value}')
```

#### Since p-value in ADH test is more than significance level of 0.05, the null hypothesis is accepted. Thus, the time series possesses a unit root and is non-stationary.

```python
# 1st differencing to make the data stationary
covidData_Northeast['pos_diff'] = pos_per_capita_Northeast - pos_per_capita_Northeast.shift(1)
covidData_Northeast['pos_diff'].dropna().plot()
```

```python
covidData_Northeast = covidData_Northeast.drop(pos_per_capita_Northeast,axis=1)
covidData_Northeast.head()
```

```python
plot_df(covidData_Northeast, x = covidData_Northeast.date, y = covidData_Northeast.pos_diff, title = 'Northeast Case Count Per Capita after 1st differencing')
```

```python
#ADF Test
from statsmodels.tsa.stattools import adfuller
pos_diff_Northeast = pd.DataFrame(covidData_Northeast['pos_diff']) 
result = adfuller(pos_diff_Northeast.values[1:12], autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'{key}, {value}')
```

#### Since p-value in ADH test is now less than significance level of 0.05, the null hypothesis is rejected. Thus, the time series is stationary.

```python
from scipy import signal
detrended = signal.detrend(covidData_Northeast.pos_diff[1:12].values)
plt.plot(detrended)
plt.title('Northeast Case Count Per Capita Detrended by Dubtracting the Least Squares Fit', fontsize=16)
```
